```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline
```

```python
model = "deepset/roberta-base-squad2"
# all the pretrained ,odel will have 2 bodies  1. tokenization 2. learning on the tokenized data.

tokenizer = AutoTokenizer.from_pretrained(model)

model = AutoModelForQuestionAnswering.from_pretrained(model)
```

```python
qanda = pipeline("question-answering",model=model,tokenizer=tokenizer)

#from this model we have slice our tokenizer which is acually doing the encoding or victorziation process,
#from this model your going to slice the model-out, which is actually doing the training on the data
#then your doing to build the piple line of what task which model and what is your tokenizer
```

```python
# We should have as Data,lets come up with a scenario where we can have
# Corpus - its like whole good data, everypage in book is like a document , lets build a Q&A system on this

corpus =["Oranges are a widely enjoyed citrus fruit known for their bright color sweet and juicy flesh",
         " and abundance of vitamin c. They are a good source of"
         "various nutrients and are consumed in various forms, including fresh, as"
         "juice, or in other prepared foods like jams and marmalades. Orange trees,"
         "originally from Southeast Asia, are now cultivated in many warm regions globally"]
```

```python
# as the user to give any input , user input and write the prompt with the help of input statment
# lets try to ask use to ask the Question
# tell the program that if your trying to have some random data, if you come up with word called exit break the program.
# send that input to the model.QnA let it fetch the answer from this,this is how we write the syntax
# what we'll tell the program to break the loop.
# if not what it should do , we will try to form a response kind of template
# for a model send the question and the context use_input acts like a question, corpus which we have writen its acting like a context.
# Model will see the Question ! and model will see the corpus and try to give us an answer !
# that how we need to write a logic
```

```python
while True:
  user_input = input("Ask a question,please press 'exit', to stop the loop")
  if user_input.lower() == "exit":
    break
  # These lines should be at the same indentation level as the 'if' statement
  input_to_model={
        "question": user_input,
        "context" : " ".join(corpus)
    }
  response = qanda(input_to_model)
  print("Answer", response['answer'])
```
