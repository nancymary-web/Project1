```python
import fitz # PyMupdf for PDF text extraction
import torch
from transformers import pipeline
from sentence_transformers import SentenceTransformer, util
```

```python
# deepset/roberta-base-squad2
# load the hugging face QA model
qa_model = pipeline ("question-answering",model="deepset/roberta-base-squad2")
```

```python
# This is the main model take the quetsion annalize and give the answer
# what ever the data This is the main model which will understand question which will underatnd the context and give me the answer
# question in the form of the vector , text is in the form of vector
# we need to create the vector, we will load the another model

```

```python
#load a sentence transformer model for creating the embeddings/ vectors
embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
#write a small function to extract text from PDF
def extact_text_from_pdf(pdf_path):
  doc = fitz.open(pdf_path)
  text =[page.get_text("text") for page in doc]
  return text
```

```python
# function to get the most relevent text_chunk using the embeddings and
# code for cosine similarity
# function to get the most relevent text_chunk using the embeddings and
# code for cosine similarity
def get_most_relevant_text_chunk(question, text_chunks):
  question_embedding = embedding_model.encode(question,covert_to_tensor =True)
  chunk_embeddings = [embedding_model.encode(chunk,convert_to_tensor = True) for chunk in text_chunks]

  similar = [util.pytorch_cos_sim(question_embedding, chunk_embedding)[0].item() for chunk_embedding in chunk_embeddings]

  best_chunk_index = similar.index(max(similar))

  return text_chunks[best_chunk_index]

```

```python
#main q&a function
def-q_a(query,pdf_path):
  text_chunks = extact_text_from_pdf(pdf_path) # extract and split the test
  most_relevant_chunk = get_most_relevant_text_chunk(query,text_chunks) # find the best chunk
  result = qa_model(question=query,context=most_relevant_chunk) # use QA model to extract the answer
  return result
```

```python
pdf_path = '/content/Lakshmi_Resume_Updated (1).pdf'
question  = input('Enter the question')
answer = q_a(question,pdf_path)
print(f"Answer: {answer}")
```
